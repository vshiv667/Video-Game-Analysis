# 4.Creating a custom function

```python
def vivek_fn(df,category=None,value=None,n=10): #mention the default values for function arguments
    if category!=None and value!=None:
        df2=df[df[category]==value]
        return df2.sort_values(axis=0,by="Global_Sales",ascending=False,inplace=False).head(n)
    else:
        return df.sort_values(axis=0,by="Global_Sales",ascending=False,inplace=False).head(n)
```

# 5.Let's use our function to display required tables

# Top 5 Overall sellers globally

```python
vivek_fn(Vivekdf,n=5) #call function
```

# Top 10 Wii Sellers

```python
vivek_fn(Vivekdf,"Platform","Wii")
```

# Top 5 selling puzzle games

```python
vivek_fn(Vivekdf,"Genre","Puzzle",5)
```

# Top 10 selling games in 1990

```python
vivek_fn(Vivekdf,"Year",1990)
```
